### Fetching Data from Feelings Section

```python
def GetDataInFrame(tag,rows_to_get,link):
    ls = []
    data = requests.get(link)
    soup = BeautifulSoup(data.text,'lxml')
    find_tag = soup.find_all(name=tag)[:rows_to_get]
    ls = [text.text for text in find_tag]
    return ls
```

```python
poetry_data = {}
```

```python

data_to_scrape = [['dua','mohabbat','sad','bewafa','funny','miss-you','tanhai'],
                  ['hamd','naat','aqwal-e-zareen','sufiana'],
                  ['new-year','ramzan','rabi-ul-awal','eid','barish'],
                  ['punjabi-ghazals','urdu-ghazals'],
                  ['urdu','punjabi','english'],
                  ['ahmed-faraz','allama-iqbal','altaf-hussain-hali','amjad-islam-imjad','anwar-masood','faiz-ahmed-faiz',
                   'mirza-ghalib','sagir-siddiqui','warish-shah','wasi-shah']]
Sections = ['feelings','islamic','events','ghazals','language','poets']
for index,sect in enumerate(Sections):
    print("\t\t Getting Data from {} Section".format(sect))
    for ctgry in data_to_scrape[index]:
        result = GetDataInFrame('p',8,'http://adaabeishq.com/category/feelings/{}/'.format(ctgry))
        '''
        Parameters:
        p ==> to get all data from <p> tag
        8 ==> to fetch only the top 8 values
        https://adaabeishq.com/...feelings ==> link to scrape the data from
        '''
        col = '{}({})'.format(sect,ctgry)
        poetry_data[col] = [x for x in result]
        print("Scraped from {} Category".format(ctgry))
```
